```python
#6.7
def max_value(lst):
  max_atual = None
  for i in range(0,len(lst)-1):
      max_atual = max(lst[i], lst[i+1])
  return max_atual

print(max_value([1,2,3,4,5]))
```

```python
#6.9
def factorial(n):
  res =1
  for i in range(1, n+1):
      res *= i
  return res

print(factorial(5))
```

```python
#6.10
def fibonnaci(n):
    a = 0
    b = 1
    if n < 0:
        print("so existe fibonnci de numeros maiores que 0")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for i in range(2, n+1):
            c = a + b
            a = b
            b = c
        return b

fibonnaci(9)
```
