```python
def slice_and_print(input_text, chunk_size):
    # Dividindo a frase em substrings do tamanho especificado
    substrings = [input_text[i:i+chunk_size] for i in range(0, len(input_text), chunk_size)]

    # Exibindo cada substring em uma linha
    for substring in substrings:
        print(substring)

def main():
    user_input = input("Digite uma frase: ")
    chunk_size = 6  # Tamanho desejado para as substrings
    slice_and_print(user_input, chunk_size)

if __name__ == "__main__":
    main()
```
