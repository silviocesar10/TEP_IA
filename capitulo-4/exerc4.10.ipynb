```python
#frase = input("Digite uma frase com artigos: ")
#nf = frase.lower().replace(" o "," ").replace(" os "," ").replace(" a "," ").replace(" as "," ")
#nf
#print(nf)
import re

def remove_articles(input_text):
    # Lista de artigos em português
    articles = ["o", "a", "os", "as", "um", "uma", "uns", "umas"]

    # Padronizando a entrada para minúsculas para fazer a correspondência de forma case-insensitive
    input_text_lower = input_text.lower()

    # Definindo a expressão regular para encontrar os artigos
    articles_pattern = r'\b(' + '|'.join(articles) + r')\b'

    # Substituindo os artigos na frase com um espaço em branco
    text_without_articles = re.sub(articles_pattern, '', input_text_lower)

    return text_without_articles

def main():
    user_input = input("Digite uma frase: ")
    result = remove_articles(user_input)
    print("Frase sem os artigos:", result)

main()
```
