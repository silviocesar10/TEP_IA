```python
#7.3
f1 = "arquivo.txt"
f2 = "resultado7_3.txt"
fpth = "sample_data/" + f1
fpth2 = "sample_data/" + f2
f = open (fpth, "r")
r = open (fpth2, "w")
while l := f.readline():
    r.write(l[::-1].replace("\n", "") + "\n")
r.close()
r = open (fpth2, "r")
while rt := r.readline():
    print(rt)
r.close()
```
