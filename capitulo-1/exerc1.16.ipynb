```python
valor = None
taxa = None
anos = None
valor = float(input("Digite o valor inicial do investimento: \n"))
taxa = float(input("Digite o valor da taxa de juros: \n"))
anos = int(input("Digite a quantidade de anos deste investimento: \n"))
resultado = valor * ((1 + (taxa / 100))**anos)
print(f"O valor pago no final sera de : {resultado:1.2f}")
```
