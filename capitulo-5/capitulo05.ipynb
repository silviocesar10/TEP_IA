```python
#5.11
def eh_par(n):
  if(n % 2 ==0):
    return n

n = [int(x) for x in input("Digite uma lista de números: ").split()]
l = list(map(eh_par, n))
for i in l:
  if(i == None):
    l.remove(i)

print(f"Lista de numeros pares {l}")
```

```python
#5.13
lc = {}
la = input("Digite 5 numeros separados por espaco: ")
lb = input("Digite 5 outros numeros: ")
la = la.split()
lb = lb.split()

for a, b in zip(la, lb):
    a = float(a)
    b = float(b)
    media = (a + b) / 2
    lc[(a, b)] = media

print(lc)
```

```python
#5.14
la = input("Digite uma lista de nome de alunos:")
lb = input("Digite uma lista de suas respectivas notas:")
la = la.split()
lb = lb.split()

lc = list(zip(la, lb))
lc.sort(key=lambda x: int(x[1]), reverse=True)
print("Lista de tuplas em ordem decrescente de nota:")
for al, nt in lc:
    print(f"Aluno: {al}, Nota: {nt}")
```

```python
#5.15
l = input("Digite uma lista de números: ")
r = int(input("Digite o número de referência: "))
nmrs = [int(x) for x in l.split()]

encontrei = False
for i, num in enumerate(nmrs):
    if num > r:
        print(f"O primeiro número maior que {r} está na posição {i}.")
        encontrei = True
        break

if not encontrei:
    print(f"Não existe um número maior que a referência {r}.")
```

```python
#5.16
l = input("Digite uma lista de números: ")
nl = [int(x) for x in l.split()]
inv_lst = [x for i,x in enumerate(nl) if x % 2 == 0][::-1]
for i in nl:
    for x in inv_lst:
        if i == x:
            continue
        else:
            if inv_lst.__contains__(i):
                continue
            else:
                inv_lst.append(i)
print(inv_lst)
```
